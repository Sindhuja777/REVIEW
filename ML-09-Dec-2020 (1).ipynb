```python
import pandas as pd
```

```python
automob_df = pd.read_csv("Datasets/automobile.csv")
automob_df.head()
```

```python
automob_df.isnull().sum()
```

```python
automob_df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
automob_df['make'] = le.fit_transform(automob_df['make'])
```

```python
automob_df['fuel-type'] = le.fit_transform(automob_df['fuel-type'])
```

```python
automob_df['num-of-doors'] = le.fit_transform(automob_df['num-of-doors'])
```

```python
automob_df['body-style'] = le.fit_transform(automob_df['body-style'])
```

```python
automob_df['engine-location'] = le.fit_transform(automob_df['engine-location'])
```

```python
automob_df['num-of-cylinders'] = le.fit_transform(automob_df['num-of-cylinders'])
```

```python
automob_df['horsepower'] = le.fit_transform(automob_df['horsepower'])
```

```python
automob_df['peak-rpm'] = le.fit_transform(automob_df['peak-rpm'])
```

```python
automob_df.head()
```

```python
automob_df.info()
```

```python
am_corr = automob_df.corr()
am_corr[am_corr['price']>0]
```

```python
automob_df_final = automob_df[['engine-location','length',
                              'width','num-of-cylinders',
                              'price']]

automob_df_final.head()
```

```python
X = automob_df_final[['engine-location','length','width',
                     'num-of-cylinders']]

y = automob_df_final['price']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor()
```

```python
dtr.fit(X_train,y_train)
```

```python
dtr.score(X_train,y_train)
```

```python
y_pred = dtr.predict(X_test)
```

```python
dtr.score(X_test,y_test)
```

#### Random Forest Regressor

```python
house_df = pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Regression/kc_house_data.csv")
house_df.head()
```

```python
house_df.isnull().sum()
```

```python
house_df.info()
```

```python
house_df.drop(['id','date'],axis=1,inplace=True)
```

```python
house_df.corr()
```

```python
X = house_df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
       'waterfront', 'view', 'condition', 'grade', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long',
       'sqft_living15', 'sqft_lot15']]

y = house_df['price']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train,y_train)
```

```python
rfr.score(X_train,y_train)
```

```python
rfr.predict(X_test)
```

```python
rfr.score(X_test,y_test)
```
